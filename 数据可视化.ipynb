```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
ts=pd.Series(np.random.randn(1000),index=pd.date_range('2000/1/1',periods=1000))
ts=ts.cumsum()
ts.describe()
```

```python
ts.plot()#线图
```

```python
df=pd.DataFrame(np.random.randn(1000,4),index=ts.index,columns=list('ABCD'))
df=df.cumsum()
df.describe()
```

```python
df.iloc[0].plot(kind='bar')#
```

```python
df['ID']=np.arange(len(df))
df.plot(x='ID',y=['A','C'])
```

```python
df.plot.bar()
```
